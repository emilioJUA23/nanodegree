```python
import psycopg2
```

```python
try: 
    conn = psycopg2.connect("host=127.0.0.1 port=7000 dbname=postgres user=postgres password=example")
except psycopg2.Error as e: 
    print("Error: Could not make connection to the Postgres database")
    print(e)
try: 
    cur = conn.cursor()
except psycopg2.Error as e: 
    print("Error: Could not get cursor to the Database")
    print(e)
conn.set_session(autocommit=True)
```

```python
# TO-DO: Add the CREATE Table Statement and INSERT statements to add the data in the table

try: 
    cur.execute("CREATE TABLE IF NOT EXISTS music_store (transaction_id int, customer_name varchar, cashier_name varchar, year int, albums_puchased varchar)")
except psycopg2.Error as e: 
    print("Error: Issue creating table")
    print (e)
    
```
